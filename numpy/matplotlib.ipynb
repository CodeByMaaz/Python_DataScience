```python
days = [1,2,3,4,5]
sleeping = [3,5,5,6,7]
eating = [2,3,6,7,8]
working = [5,3,2,6,7]
playing = [23,5,6,23,4]
plt.plot([],[], color = 'm',label ='sleeping', linewidth = 5)
plt.plot([],[], color = 'c',label ='Eating', linewidth = 5)
plt.plot([],[], color = 'r',label ='walking', linewidth = 5)
plt.plot([],[], color = 'k',label ='playing', linewidth = 5)
plt.stackplot(days, sleeping, eating, working, playing, colors = ['m','c','r','k'])
plt.xlabel('x axis')
plt.ylabel('y label')
plt.legend()
plt.show()
```

```python
slices = [7,2,5,23]
activities = ['sleeping','eating', 'walking', 'playing']
cols = ['c','m','g','r']
plt.pie(slices,
       labels = activities,
        colors = cols,
        startangle = 90,
        shadow = True,
        explode = (0,0.1,0,0),
        autopct = '%1.1f%%'
        
       )
plt.title('pie plot')
plt.show()
```
